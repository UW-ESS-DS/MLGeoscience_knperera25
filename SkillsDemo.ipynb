```python
##Demonstration of Python Skills

# Importing necessary modules
import numpy as np
import matplotlib.pyplot as plt

# Declaring a list
my_list = [1, 2, 3, 4, 5]

# Declaring a NumPy array
my_array = np.array([1, 2, 3, 4, 5])

# Defining a basic Python function
def square_numbers(numbers):
    squared_nums = [num ** 2 for num in numbers]
    return squared_nums

# Using the function with the list
squared_list = square_numbers(my_list)

# Using the function with the NumPy array
squared_array = square_numbers(my_array)

# Creating a simple Matplotlib figure
plt.figure(figsize=(6, 4))
plt.plot(my_list, squared_list, label='List Squares', marker='o')
plt.plot(my_array, squared_array, label='Array Squares', marker='o')
plt.xlabel('Numbers')
plt.ylabel('Squared Values')
plt.title('Squared Numbers')
plt.legend()
plt.grid(True)

# Saving the figure as a PNG file
plt.savefig('squared_numbers_plot.png')

# Displaying the plot (optional, if you are working in an interactive environment)
plt.show()
```
